# Merge and Transform Data

```python
fundamentals_df['date'] = pd.to_datetime(fundamentals_df['date'], format='%Y-%m-%d')
prices_df['date'] = pd.to_datetime(prices_df['date'], format='%Y-%m-%d')
#securities_df['date'] = pd.to_datetime(securities_df['date'], format='%Y-%m-%d')
#fundamentals_df['date']

#About 25% of fundamental_df dates do not match prices_df dates. Therefore use merge_asof() to 
#find and merge all dates that have not merged. Only needed a search to within 4 days to merge 99% of dates.

stocks_df = pd.merge_asof(fundamentals_df.sort_values('date'), prices_df.sort_values('date'), on='date', by='symbol', tolerance=pd.Timedelta(days=4),allow_exact_matches=False)
stocks_df = stocks_df.merge(securities_df, on=['symbol'],how='left')
stocks_df.count()

#type(new[0])
```

```python
stocks_df = stocks_df.sort_values('symbol', ascending=True, na_position='first')
stocks_df = stocks_df.loc[(stocks_df['date'] > '2011-01-01')]

new_names = ['date','symbol','Security','GICS Sub Industry','close', 'volume', 'Total Revenue', 'Net Income','Long-Term Debt', 'Capital Expenditures']
stocks_df = stocks_df[new_names].copy()
stocks_df.reset_index(drop=True, inplace=True)
stocks_df.head(10)
```
